```python
import numpy as np
import pandas as pd
import dateutil.parser
import matplotlib.pyplot as plt
from collections import Counter
%matplotlib notebook
```

```python
f = open('E:/Keylogs/Typing_Participants/Participant_7_20171213/Subject_mperf_alabsi_7/keyboardlogs0.txt', 'r')
x = f.readlines()
f.close()

time = []
keys = []
active_window = []
key_ids = []

for line in x[2:]:
	if not line.startswith(("//"," ","\n")):
		line = line.strip('\n')
		line = line.split(',')
		time.append(line[0].strip())
		keys.append(line[1].strip())
		active_window.append(line[2].strip())
		key_ids.append(line[3].strip())

n = len(key_ids)
print (time[0])
print (keys)
print (active_window)
print (key_ids)
```

```python
l = len(keys)
for i in range(0,l):
    if not key_ids[i].isdigit():
        key_ids[i] = 100
Counter(key_ids)    
```

```python
time_utc = []
for i in range(0,l): 
    yourdate = dateutil.parser.parse(time[i])
    time_utc.append(yourdate.timestamp() - 7200) #timezone correction for 2 hours
time_utc = np.array(time_utc) * 1000
print(len(keys))
print(len(key_ids))
print(len(time))
print(len(time_utc))
```

```python
start_time = time_utc[0]
end_time = time_utc[-1]
print("Total time in mins is", (end_time - start_time)/ 60000)
delta = 40 
constant_t = (np.arange(start_time, end_time , delta))
print(constant_t)
```

```python
labels = pd.DataFrame(0, index=np.arange(len(constant_t)), columns=['Time' , 'Key_id', 'Key'])
labels['Time']  = constant_t
labels
```

```python
time2 = time_utc - start_time

for i in range(0,l):
    ind = int(np.floor((time2[i])/40))
    labels.loc[ind, 'Key_id'] = key_ids[i]
    labels.loc[ind, 'Key'] = keys[i]
print(labels)
```

```python
labels_np = np.zeros((len(constant_t),3))
labels_np[:,0] = constant_t
time2 = time_utc - start_time

for i in range(0,l):
    index = int(np.floor((time2[i])/40))
    labels_np[index,1] = key_ids[i]
    labels_np[index,2] = 1
```

```python
print(labels_np.shape)
print(labels_np)
```

```python
np.save('Participant_7_label.npy', labels_np)
```
